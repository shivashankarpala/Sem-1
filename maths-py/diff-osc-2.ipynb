```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def f(u, x):
        return (u[1], -(1/9)*(1.2*u[1] + 2*u[0]))

y0 = [2.5, 1.5]
xs = np.linspace(0, 20*np.pi, 2000)
us = odeint(f, y0, xs)
ys = us[:, 0]

print(us)
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("Solution of Free and Undamed case")
plt.plot(xs, ys, 'r-')
plt.grid()
plt.show()
```
