```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def dudx(u, x):
        return (u[1], -2*u[1] + 2*u[0] + np.cos(2*x))

u0 = [0, 0]
xs = np.linspace(1, 10, 200)
us = odeint(dudx, u0, xs)
ys = us[:, 0]

plt.xlabel("t values")
plt.ylabel("x values")
plt.title("Solution Curves")
plt.plot(xs, ys, 'r-')
plt.grid()
plt.show()
```
