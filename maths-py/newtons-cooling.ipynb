```python
import numpy as np
from sympy import *
import matplotlib.pyplot as plt

t2=20
t1=100
t=10
T=75

k1=(1/t)*log((t1-t2)/(T-t2))
print('k= ',k1)

k= Symbol('k')
t= Symbol('t')
T=Function('T')(t)
T=t2+(t1-t2)*exp(-k*t)
print('T= ',T)

T=T.subs(k,k1)
T=lambdify(t,T)
t=np.linspace(0,70)

plt.plot(t, T(t), color='red')
plt.grid()
plt.show()
print('When time t=30 minute T is,',T(30),'o C')
```
