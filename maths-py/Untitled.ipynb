```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def f(u, x):
        return (u[1], -36*u[0])

y0 = [0, 1]
xs = np.linspace(0, 5, 10)
us = odeint(f, y0, xs)
ys = us[:, 0]

print(ys)
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("Solution of Free and Undamed case")
plt.plot(xs, ys, 'r-')
plt.grid()
plt.show()
```
