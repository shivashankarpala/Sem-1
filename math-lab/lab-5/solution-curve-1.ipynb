```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

Kp=3
taup=2

def model(y,t):
    u = 1
    return (-y + Kp*u)/taup

t3=np.linspace(0,14,100)
y3=odeint(model,0,t3)

plt.plot(t3,y3, 'r-', linewidth=1, label='Ode Integrator')
plt.xlabel('time')
plt.ylabel('Respomse(y)')
plt.legend(loc='best')
plt.show()
```
